# Sigmoid function 

```python
import numpy as np
import matplotlib.pyplot as plt 
import math

%matplotlib inline 
```

```python
def sigmoid(x):
    g = []
    for num in x:
        g.append(1 / (1 + math.exp(-num)))
    return g
```

```python
x = np.arange(-10, 10, 0.1)
sig = sigmoid(x)
```

```python
plt.plot(x, sig)
plt.title('Sigmoid function')
```
